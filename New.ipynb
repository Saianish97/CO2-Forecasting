# Data Cleaning

```python
import re # Regular expression
import string

def clean_train(train):
    ''' Make text lowercase, remove text in sqaure brackets, remove punctuation and remove wprds containing numbers.'''
    train = train.lower()
    train = re.sub('\[.*?\]', '', train)
    train = re.sub('[%s]' % re.escape(string.punctuation), '', train)
    train = re.sub('\w*\d\w*', '', train)
    train = re.sub("[0-9" "]+"," ",train)
    train = re.sub('[‘’“”…]', '', train)
    return train

clean = lambda x: clean_train(x)
```

```python
clean("This is a SalaryData #$345% Train%$#@ 987")
```

```python
data['Age'] = data.age.apply(clean)
data.age
```
