```python
# Numerical Variables
numerical_features=[feature for feature in forestfires.columns if forestfires[feature].dtypes != '0']

print('Number of numerical variables:', len(numerical_features))

# Visulaization the variables
forestfires[numerical_features].head()
```

```python
discrete_feature=[feature for feature in numerical_features if len(forestfires[feature].unique())<25]
print('Discrete Variables Count: {}'.format(len(discrete_feature)))
```

```python
# Continuous Variable
continuous_feature=[feature for feature in numerical_features if feature not in discrete_feature]
print('Continuous Feature Count {}'.format(len(continuous_feature)))
```

```python
# Categorical Variables
categorical = [var for var in forestfires.columns if forestfires[var].dtype=='0']
print('There are {} categorical variables\n'.format(len(categorical)))
print('The categorical variables are :\n\n', categorical)
```
